```python
#import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the dataset -sns
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
data = load_iris()
data
```

```python
#feauture and target selection
x = data.data
y = data.target
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtr, xte, ytr, yte = train_test_split(x,y,test_size=0.2,random_state=16)
```

```python
xtr.shape
```

```python
ytr.shape
```

```python
xtr
```

```python
#standardization -scaling
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
xtr_scaled = ss.fit_transform(xtr)
xtr_scaled
```

```python
#algo - logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
#instantiate the LR class
lr_model = LogisticRegression()
lr_model.fit(xtr_scaled,ytr)
```

```python
#prediction 
lr_pred = lr_model.predict(xte)
```

```python
lr_pred
```

```python
#evaluation
acc = accuracy_score(yte, lr_pred)
print(f'The accauracy score is {acc*100:.2f}%')
```

```python
#confusion matrix
cm = confusion_matrix(yte, lr_pred)
cm
```

```python
#classification report
cr = classification_report(yte, lr_pred)
print(cr)
```
